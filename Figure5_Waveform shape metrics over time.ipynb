<b>Calculate waveform metrics in windows

We use a 30 second sliding window with 5 second overlap to show how each metric changes throughout a recording.

```python
group='C'
subject=2
sWindow=30#window length in seconds
sSlide=5#slide length in seconds

eeg,rejects = utils.loadPD()
PACseg,ShRseg,StRseg=utils.metricsegment(eeg,rejects,group,subject,sWindow,sSlide)
```

Each metric fluctuates throughout the recording session. 

```python
font_=18
uppery=max(max(ShRseg),max(PACseg),max(StRseg))
y_lim=(0,uppery+.02)
c1='lightcoral'
c2='steelblue'
c3='lightseagreen'
#create figure
fig=plt.figure(figsize=(10,2.5))
plt.ylim(y_lim)
plt.plot(ShRseg,label='ShR',color=c1)
plt.ylim(y_lim)
plt.plot(PACseg,label='PAC',color=c2)
plt.ylim(y_lim)
plt.plot(StRseg,label='StR',color=c3)
plt.legend(loc='best',fontsize=12)
plt.xlim(0,len(ShRseg)-1)
plt.ylim(y_lim)
plt.xlabel('Window Number',fontsize=font_)
#plt.savefig('Figtrace.eps', format='eps', dpi=1000)
plt.show()
plt.tight_layout()
```
